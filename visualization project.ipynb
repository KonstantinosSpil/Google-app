# Start cleaning

### Convert Size to MB and float

```python
df['Size'] = df['Size'].str.replace(',', '.', regex=False)
```

```python
# function to convert app size to float and only to MB
def size_to_float(x):
    if isinstance(x, float):
        return x  # If already a float, return as is
    
    if x.endswith('k'):
        x = float(x[:-1]) / 1024
        return x
    elif x.endswith('M'):
        x = float(x[:-1])
        return x
    elif x.endswith('G'):
        x = float(x[:-1]) * 1024
        return x
    elif x == "Varies with device":
        return np.nan
    
    return x
```

```python
df["Size"] = df["Size"].apply(size_to_float)
```

### creating a new column and converting all prices to EUR

```python
df["Currency"] = df["Currency"].replace("XXX", np.nan)
```

```python
df["Currency"].unique()
```

```python
def currency_to_EUR(price, currency):
    
    exchange_rates = {
        'USD': 1.12,
        'CAD': 0.68,
        'EUR': 1.0, 
        'INR': 0.012,
        'VND': 4.5e-05,
        'GBP': 1.16,
        'BRL': 0.15,
        'KRW': 0.00074,
        'TRY': 0.11,
        'RUB': 0.011,
        'SGD': 0.62,
        'AUD': 0.64,
        'PKR': 0.0057,
        'ZAR': 0.061
    }
    
    if currency in exchange_rates:
        exchange_rate = exchange_rates[currency]
        return price * exchange_rate
    else:
        return None
```

```python
df['Price_EUR'] = df.apply(lambda row: currency_to_EUR(row['Price'], row['Currency']), axis=1)
```

```python
df["Price_EUR"] = df["Price_EUR"].round(2)
```
